# Perhitungan Section Modulus Midship Section


---

## 1. Import Library & Baca File CSV

```python
import pandas as pd
import numpy as np

pd.set_option('display.float_format', '{:.4f}'.format)
pd.set_option('display.max_rows', 200)
pd.set_option('display.max_columns', 20)
pd.set_option('display.width', 200)

# Baca file CSV (separator semicolon)
df = pd.read_csv('scantlling_rapi.csv', sep=';')

# Rename kolom agar lebih mudah diakses
df.columns = ['Nama Bagian', 'n', 'b', 'h', 'a', 'Z']

# Pastikan kolom numerik
for col in ['n', 'b', 'h', 'a', 'Z']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

print(f'Total elemen: {len(df)}')
print()
df.head(10)
```

---
## 2. Perhitungan Kolom Lanjutan

| Kolom | Rumus |
|-------|-------|
| **A_T** | `n × b × h` |
| **S** (Momen Luas) | `A_T × Z` |
| **S2** (Momen Inersia) | `A_T × Z²` |
| **Ix** | `(1/12) × b × h³ × cos²(a)` |
| **Iy** | `(1/12) × b × h³ × sin²(a)` |
| **I_ind** | `Ix + Iy` |

```python
# Konversi sudut dari derajat ke radian
a_rad = np.radians(df['a'])

# Luas Total: A_T = n x b x h
df['A_T (cm²)'] = df['n'] * df['b'] * df['h']

# Momen Luas: S = A_T x Z
df['S (cm³)'] = df['A_T (cm²)'] * df['Z']

# Momen Inersia terhadap base: S2 = A_T x Z²
df['S2 (cm⁴)'] = df['A_T (cm²)'] * df['Z']**2

# Momen Inersia Individu Ix = 1/12 x n x b x h³ x cos²(a)
# df['Ix (cm⁴)'] = (1/12) * df['n'] * df['b'] * df['h']**3 * np.cos(a_rad)**2
df['Ix (cm⁴)'] = (1/12) * df['b'] * df['h']**3 * np.cos(a_rad)**2

# Momen Inersia Individu Iy = 1/12 x n x b x h³ x sin²(a)
# df['Iy (cm⁴)'] = (1/12) * df['n'] * df['b'] * df['h']**3 * np.sin(a_rad)**2
df['Iy (cm⁴)'] = (1/12) * df['b'] * df['h']**3 * np.sin(a_rad)**2

# Momen Inersia Individu Total
df['I_ind (cm⁴)'] = df['Ix (cm⁴)'] + df['Iy (cm⁴)']

print('Kolom lanjutan berhasil dihitung.')
df[['Nama Bagian', 'n', 'b', 'h', 'a', 'Z',
    'A_T (cm²)', 'S (cm³)', 'S2 (cm⁴)', 'Ix (cm⁴)', 'Iy (cm⁴)', 'I_ind (cm⁴)']].head(10)
```

---
## 3. Tampilkan Tabel Lengkap Semua Elemen

```python
cols_display = [
    'Nama Bagian', 'n', 'b', 'h', 'a', 'Z',
    'A_T (cm²)', 'S (cm³)', 'S2 (cm⁴)', 'Ix (cm⁴)', 'Iy (cm⁴)', 'I_ind (cm⁴)'
]

df[cols_display].style \
    .set_caption('Tabel Scantling Elemen Konstruksi Kapal') \
    .format({
        'n': '{:.0f}', 'b': '{:.2f}', 'h': '{:.2f}', 'a': '{:.1f}', 'Z': '{:.4f}',
        'A_T (cm²)': '{:.4f}', 'S (cm³)': '{:.4f}', 'S2 (cm⁴)': '{:.4f}',
        'Ix (cm⁴)': '{:.4f}', 'Iy (cm⁴)': '{:.4f}', 'I_ind (cm⁴)': '{:.4f}'
    }) \
    .set_table_styles([{
        'selector': 'th',
        'props': [('background-color', '#2c3e50'), ('color', 'white'),
                  ('font-weight', 'bold'), ('text-align', 'center')]
    }, {
        'selector': 'td',
        'props': [('text-align', 'right'), ('padding', '4px 8px')]
    }, {
        'selector': 'td:first-child',
        'props': [('text-align', 'left')]
    }])
```

---
## 4. Sigma (Penjumlahan Global)

```python
Sigma1 = df['A_T (cm²)'].sum()    # Total Luas
Sigma2 = df['S (cm³)'].sum()      # Total Momen Luas
Sigma3 = df['S2 (cm⁴)'].sum()     # Total Momen Inersia terhadap base
Sigma4 = df['I_ind (cm⁴)'].sum()  # Total Momen Inersia Individu

print('=' * 45)
print('          REKAPITULASI SIGMA')
print('=' * 45)
print(f'  Σ A_T  (Luas Total)           = {Sigma1:>12.4f} cm²')
print(f'  Σ S    (Momen Luas)           = {Sigma2:>12.4f} cm³')
print(f'  Σ S²   (Momen Inersia Base)   = {Sigma3:>12.4f} cm⁴')
print(f'  Σ I_ind (M. Inersia Individu) = {Sigma4:>12.4f} cm⁴')
print('=' * 45)
```

---
## 5. Perhitungan Neutral Axis (Sumbu Netral)

```python
H_kapal = 609.6  # cm  (tinggi kapal)

# Titik berat terhadap bottom
Z1 = Sigma2 / Sigma1

# Titik berat terhadap deck
Z2 = H_kapal - Z1

print('=' * 45)
print('          NEUTRAL AXIS')
print('=' * 45)
print(f'  H kapal                  = {H_kapal:>10.4f} cm')
print(f'  Z₁ (dari baseline/keel)  = {Z1:>10.4f} cm')
print(f'  Z₂ (dari deck)           = {Z2:>10.4f} cm')
print('=' * 45)
```

---
## 6. Perhitungan Momen Inersia terhadap Sumbu Netral (I_NA)

```python
# Momen inersia total terhadap baseline
Ixx = Sigma3 + Sigma4

# Momen inersia terhadap sumbu netral (Steiner theorem)
I_NA = Ixx - (Z1**2) * Sigma1

print('=' * 55)
print('     MOMEN INERSIA TERHADAP SUMBU NETRAL')
print('=' * 55)
print(f'  Ixx  = Σ S² + Σ I_ind     = {Ixx:>14.4f} cm⁴')
print(f'  I_NA = Ixx - Z₁² × Σ A_T  = {I_NA:>14.4f} cm⁴')
print('=' * 55)
```

---
## 7. Kalkulasi Section Modulus

```python
# Modulus penampang (cm³)
W_bot_cm = I_NA / Z1
W_deck_cm = I_NA / Z2

# Konversi ke m³ (1 m³ = 10⁶ cm³)
W_bot_m = W_bot_cm / 1e6
W_deck_m = W_deck_cm / 1e6

print('=' * 55)
print('           SECTION MODULUS')
print('=' * 55)
print(f'  W_bot  = I_NA / Z₁  = {W_bot_cm:>14.4f} cm³')
print(f'                      = {W_bot_m:>14.6f} m³')
print()
print(f'  W_deck = I_NA / Z₂  = {W_deck_cm:>14.4f} cm³')
print(f'                      = {W_deck_m:>14.6f} m³')
print('=' * 55)
```

---
## 8. Ringkasan Hasil Akhir

```python
summary = pd.DataFrame({
    'Parameter': [
        'Tinggi Kapal (H)',
        'Σ Luas Total (Σ A_T)',
        'Σ Momen Luas (Σ S)',
        'Σ M. Inersia Base (Σ S²)',
        'Σ M. Inersia Individu (Σ I_ind)',
        'Neutral Axis dari Bottom (Z₁)',
        'Neutral Axis dari Deck (Z₂)',
        'Momen Inersia Total (Ixx)',
        'Momen Inersia NA (I_NA)',
        'Section Modulus Bottom (W_bot)',
        'Section Modulus Deck (W_deck)',
        'Section Modulus Bottom (W_bot)',
        'Section Modulus Deck (W_deck)',
    ],
    'Nilai': [
        f'{H_kapal:.4f}',
        f'{Sigma1:.4f}',
        f'{Sigma2:.4f}',
        f'{Sigma3:.4f}',
        f'{Sigma4:.4f}',
        f'{Z1:.4f}',
        f'{Z2:.4f}',
        f'{Ixx:.4f}',
        f'{I_NA:.4f}',
        f'{W_bot_cm:.4f}',
        f'{W_deck_cm:.4f}',
        f'{W_bot_m:.6f}',
        f'{W_deck_m:.6f}',
    ],
    'Satuan': [
        'cm', 'cm²', 'cm³', 'cm⁴', 'cm⁴',
        'cm', 'cm',
        'cm⁴', 'cm⁴',
        'cm³', 'cm³',
        'm³', 'm³',
    ]
})

summary.style \
    .set_caption('Ringkasan Hasil Perhitungan Section Modulus Midship') \
    .set_table_styles([{
        'selector': 'th',
        'props': [('background-color', '#2c3e50'), ('color', 'white'),
                  ('font-weight', 'bold'), ('text-align', 'center')]
    }, {
        'selector': 'tr:last-child td, tr:nth-last-child(2) td',
        'props': [('font-weight', 'bold'), ('background-color', '#eafaf1')]
    }, {
        'selector': 'td',
        'props': [('padding', '5px 12px')]
    }]) \
    .hide(axis='index')
```

## 9. Kalkulasi Kekuatan Material

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Data A36 Ultimate Strength (MPa)
data = np.array([
    355.504, 439.795, 472.011, 263.955, 376.981, 342.362,
    478.974, 484.172, 485.152,
    360.7, 400.4, 468.2, 426.9, 363.9, 391.1,
    462.3, 435.7, 355.7, 401.1, 472.0, 418.0
])

# Statistik
mean = np.mean(data)
std  = np.std(data, ddof=1)   # ddof=1 → sample std deviation
n    = len(data)

print(f"Jumlah data (n)     : {n}")
print(f"Rata-rata (μ)       : {mean:.3f} MPa")
print(f"Std. Deviasi (σ)    : {std:.3f} MPa")

# Plot
x = np.linspace(mean - 4*std, mean + 4*std, 500)
pdf = norm.pdf(x, mean, std)

fig, ax = plt.subplots(figsize=(9, 5))

# Histogram (dinormalisasi ke density)
ax.hist(data, bins=8, density=True,
        color='#5DCAA5', edgecolor='#0F6E56', alpha=0.55,
        label='Histogram data')

# Kurva PDF
ax.plot(x, pdf, color='#185FA5', linewidth=2.5,
        label=f'PDF Normal\nμ = {mean:.2f} MPa\nσ = {std:.2f} MPa')

# Garis vertikal mean
ax.axvline(mean, color='#D85A30', linestyle='--', linewidth=1.5,
           label=f'Mean = {mean:.2f} MPa')

# Label & format
ax.set_xlabel('Ultimate Strength (MPa)', fontsize=13)
ax.set_ylabel('Probability Density', fontsize=13)
ax.set_title('A36 Steel Ultimate Strength Probability Density Function', fontsize=14)
ax.legend(fontsize=11)
ax.grid(True, alpha=0.25)

plt.tight_layout()
plt.savefig('pdf_a36.png', dpi=150)
plt.show()
```
